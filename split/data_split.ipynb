```python
import os, re, glob
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from PIL import Image
import pandas as pd
```

```python
# Path는 본인 경로에 맞게 작성 부탁 드립니다!
imagePath = '../Data/Raw/train'
toPath = '../Data/train'
```

```python
categories = ["Tomato_D01", "Tomato_D04", "Tomato_D05", "Tomato_D07", "Tomato_D08", "Tomato_D09", "Tomato_H", "Tomato_P03", "Tomato_P05", "Tomato_R01"]
```

```python
nb_classes = len(categories)  
```

```python
path = []
label = []

for idx, cate in enumerate(categories):  
    tmp = 0
    image_dir = imagePath+'/'+cate+'/'
    to_dir = toPath+'/'+cate+'/'
    
    len_path = '../shared/hackathon/Raw/train' + cate
    len_file_list = os.listdir(len_path)
    len_file_train = len(len_file_list) * 0.8
    
    for top, dir, f in os.walk(image_dir): 
        for filename in f:
            df_file = len_path + '/' + filename
            print(cate + '*' + df_file)
            path.append(df_file)
            label.append(cate)
            tmp = tmp + 1
            print(tmp)
            if tmp >= len_file_train:
                img = cv2.imread(image_dir+filename)  
                cv2.imwrite(to_dir + filename, img)
```

```python
df = pd.DataFrame([ x for x in zip(path, label)])
```

```python
df.to_csv('../kwseo/augmentation/new_train_path.csv',
         index = False)
```
